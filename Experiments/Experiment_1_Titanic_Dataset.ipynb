```python
# Task 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 2: Load Titanic Dataset
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
# Task 3: Display first 10 rows
print("First 10 Rows:")
print(df.head(10))
```

```python
# Task 4: Display last 5 rows
print("\nLast 5 Rows:")
print(df.tail(5))
```

```python
# Task 5: Check dataset dimensions
print("\nDataset Shape:")
print(df.shape)
```

```python
# Task 6: Dataset Information
print("\nDataset Information:")
df.info()
```

```python
# Task 7: Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

print("\nAge Statistics")
print("Mean:", df["Age"].mean())
print("Minimum:", df["Age"].min())
print("Maximum:", df["Age"].max())

print("\nFare Statistics")
print("Mean:", df["Fare"].mean())
print("Minimum:", df["Fare"].min())
print("Maximum:", df["Fare"].max())
```

```python
# Task 8: Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Task 9: Fill missing Age values with Median
df["Age"] = df["Age"].fillna(df["Age"].median())

print("\nMissing Age values after filling:")
print(df["Age"].isnull().sum())
```

```python
# Task 10: Filter Surviving Passengers
survivors = df[df["Survived"] == 1]

print("\nFirst 5 Surviving Passengers:")
print(survivors.head())
```

```python
# Task 11: Convert Fare column to NumPy array
fare_array = df["Fare"].to_numpy()

print("\nFare NumPy Array:")
print(fare_array)
```

```python
# Task 12: Maximum and Average Fare
print("\nMaximum Fare:", np.max(fare_array))
print("Average Fare:", np.mean(fare_array))
```

```python
# Task 13: Histogram of Age
plt.figure(figsize=(6,4))
plt.hist(df["Age"], bins=20)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Task 14: Bar Chart of Survival Count
plt.figure(figsize=(5,4))
df["Survived"].value_counts().plot(kind="bar")
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
# Task 15: Scatter Plot of Age vs Fare
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Fare"])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
